```python
# Importing the packages we will use

import matplotlib.pyplot as plt  
import seaborn as sns 
import pandas as pd
```

```python
# Pre-processing

df = pd.read_csv('Pokemon.csv')

# df['Legendary'] is, originaly, a boolean which True means de pokemon is legendary and False the opposity.

df['Legendary'] = df['Legendary'].apply(lambda x: 'Not Legendary' if x == False else 'Legendary')
```

```python
df.head()
```

```python
# One Dispersion Chart

sns.scatterplot(x = 'HP', y = 'Attack', hue = 'Legendary', 
            data = df, style = 'Legendary')
sns.despine(offset = 2, trim = False)
plt.title('Dispersion Chart Example')
plt.xlabel('HP')
plt.xlim([0, 300])
plt.ylabel('Attack')
plt.ylim([0, 200])
```

```python
# Several Dispersion Charts

sns.relplot(x = 'HP', y = 'Attack', hue = 'Legendary', 
            data = df, style = 'Legendary', col = 'Generation',
            col_wrap = 3)
plt.xlabel('HP')
plt.xlim([0, 300])
plt.ylabel('Attack')
plt.ylim([0, 200])
```

```python
# Bar Chart

sns.countplot(x = 'Generation', data = df, color = 'blue')
plt.title('Bar Chart Example')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.ylim([0, 170])
```

```python
# BoxPlot chart

sns.boxplot(x = 'Legendary', y = 'HP', data = df, hue = 'Legendary', dodge=False)
plt.title('BoxPlot Chart Example')
plt.xlabel('Legendary')
plt.ylabel('HP')
```

```python
# Density chart

leg = df.HP[(df.HP) & (df.Legendary == 'Legendary')]
notleg = df.HP[(df.HP) & (df.Legendary == 'Not Legendary')]
sns.kdeplot(data = notleg, shade = True)
sns.kdeplot(data = leg, shade = True)
sns.despine(trim = False)
plt.title('Density Chard Example')
plt.xlabel('HP')
plt.ylabel('Density')
```

```python
# Histogram Chart

sns.distplot(df['HP'], kde = False, bins = 15, color = 'blue')
sns.despine(trim = False)
plt.title('Histogram Chart Example')
plt.ylabel('Count')
```
